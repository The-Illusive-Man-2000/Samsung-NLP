```python
def parse_array(s):
    return np.array([int(s.strip()) for s in s.strip().split(' ')])

def read_array():
    return parse_array(sys.stdin.readline())
```

```python
def calculate_pmi(a, b):
    #print(len(a[a * b == 1]))
    d = len(a)
    pab =len(a[a * b == 1])/d
    pa = np.count_nonzero(a)/d
    pb = np.count_nonzero(b)/d
    return np.log(pab/(pa*pb))  # ваше решение
```

```python
a = read_array()
b = read_array()
pmi_value = calculate_pmi(a, b)

print('{:.6f}'.format(pmi_value))
```
